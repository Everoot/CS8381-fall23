```python
zero_qubit = np.matrix('1; 0')
one_qubit = np.matrix('0; 1')
print(zero_qubit)
```

```python
def mix(zero, one):
    if zero + one != 1:
        raise Exception('总概率不等于100%!')
    return np.sqrt(zero) * zero_qubit + np.sqrt(one) * one_qubit 
```

```python
qubit_50_50 = mix(0.5, 0.5)
qubit_10_90 = mix(0.1, 0.9)
print(qubit_50_50)
print(qubit_10_90)
```

```python
import numpy as np
from qiskit.visualization import plot_bloch_vector
```

```python
def sph2cart(theta, phi):
    # theta: polar angle
    # phi: azimuthal angle
    x = np.sin(theta)*np.cos(phi)
    y = np.sin(theta)*np.sin(phi)
    z = np.cos(theta)
    return x,y,z
```

```python
theta = 0
phi = 0
x,y,z = sph2cart(theta,phi)
plot_bloch_vector([x,y,z])
```

$$\left | 0 \right \rangle = cos\frac{0}{2} \left | 0 \right \rangle + e^{i0}sin\frac{0}{2}\left | 1 \right \rangle$$

```python
theta = np.pi
phi = 0
x,y,z = sph2cart(theta, phi)
plot_bloch_vector([x,y,z])
```

$$ \left | 1 \right \rangle = cos\frac{\pi}{2} \left | 0 \right \rangle + e^{i0}sin\frac{\pi}{2}\left | 1 \right \rangle $$

```python
theta = np.pi/2
phi = np.pi
x,y,z = sph2cart(theta,phi)
plot_bloch_vector([x,y,z])
```
