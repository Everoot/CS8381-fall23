```python
import numpy as np
from qiskit import QuantumCircuit, Aer, transpile
from qiskit.visualization import plot_histogram
from qiskit.visualization import array_to_latex
from qiskit.providers.aer import AerSimulator
```

```python
# 创建一个三比特量子电路
qc = QuantumCircuit(3,3)

# 在q0上应用Ry门
theta = 2 * np.arccos(1/np.sqrt(3))
qc.ry(theta, 0)

# 在q1上应用H门
qc.ch(0,1)

# 使用q1作为控制比特，q2作为目标比特的CNOT门
qc.cx(1, 2)

# 使用q0作为控制比特，q1作为目标比特的CNOT门
qc.cx(0, 1)

# 使用q0作为控制比特，q2作为目标比特的CNOT门
qc.x(0)

# 可视化电路
qc.draw('mpl')

```

```python
# 使用unitary_simulator获取电路的矩阵表示
backend = Aer.get_backend('unitary_simulator')
compiled_circuit = transpile(qc, backend)
result = backend.run(compiled_circuit).result()

# 获取电路的单位矩阵
unitary_matrix = result.get_unitary()

# 使用qiskit的array_to_latex函数可视化矩阵
array_to_latex(unitary_matrix, precision=2)
```

```python
# 添加测量操作
qc.measure([0, 1, 2], [0, 1, 2])  # 将每个量子比特的测量结果存储到相应的经典比特上

# 使用Aer模拟器模拟电路
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
result = simulator.run(compiled_circuit, shots=1000).result()

# 获取模拟结果的测量统计
counts = result.get_counts()

# 可视化测量结果
plot_histogram(counts)
```
